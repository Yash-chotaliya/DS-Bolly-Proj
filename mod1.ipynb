## Import library and read csv

```python
import pandas as pd
import numpy as np

datas = pd.read_csv("D:\\DS Proj\\top 1000 bolly movies\\dataset\\base.csv")
datas
```

## drop column and change index value

```python
datas = datas.drop(["Unnamed: 0","SN"],axis=1)
datas = datas.set_index(datas.index+1)
datas
```

```python
datas[['Worldwide','India Net','India Gross','Overseas','Budget']]=datas[['Worldwide','India Net','India Gross','Overseas','Budget']].astype('float64')
datas.dtypes
```

# Replace 0 with NaN and drop null data

```python
datas = datas.replace([0,'0'],np.NaN)
datas = datas.dropna()
```

# change value in crore

```python
x = datas['Worldwide'].tolist()
y = []
for i in range(len(x)):
    y.append(x[i]/10000000)    
datas['Worldwide'] = datas['Worldwide'].replace(x,y)


x = datas['India Net'].tolist()
y = []
for i in range(len(x)):
    y.append(x[i]/10000000)    
datas['India Net'] = datas['India Net'].replace(x,y)


x = datas['India Gross'].tolist()
y = []
for i in range(len(x)):
    y.append(x[i]/10000000)    
datas['India Gross'] = datas['India Gross'].replace(x,y)


x = datas['Overseas'].tolist()
y = []
for i in range(len(x)):
    y.append(x[i]/10000000)    
datas['Overseas'] = datas['Overseas'].replace(x,y)


x = datas['Budget'].tolist()
y = []
for i in range(len(x)):
    y.append(x[i]/10000000)    
datas['Budget'] = datas['Budget'].replace(x,y)
```
